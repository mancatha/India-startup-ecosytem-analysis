### the "Amount" column contains outliers which need to be removed before moving further. Here's how to remove the outliers:

```python
# Calculate the IQR
Q1 = df['Amount($)'].quantile(0.25)
Q3 = df['Amount($)'].quantile(0.75)
IQR = Q3 - Q1

# define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# remove outliers
df = df[(df['Amount($)'] >= lower_bound) & (df['Amount($)'] <= upper_bound)]
```

```python
#Let us create a new column named 'Age'
#Below function will automate this task

def subtract_columns(df, Founded, Funding_Year, Age):
    """
    #Subtract the values in column1 from column2 row-wise and store the output in a new column.

    Parameters:
        df: The DataFrame containing the columns.
        Founded (str): The name of the first column with numerical values.
        Funding Year(str): The name of the second column with numerical values.
        Age(str): The name of the new column to store the subtraction results.

    Returns:
        df: The original DataFrame with the new column added.
    """
    df['Age'] = df['Funding_Year'] - df['Founded']
    return df
```

```python
df['Funding_Year'] = df['Funding_Year'].astype('int')
```

```python
# Call the function to create a new column ''Age'' with the results of 'Funding_Year' - 'Founded'
df = subtract_columns(df,'Funding_Year','Founded','Age')
```

```python
# Group states and calculate total funding amount BY Age
# Group by 'Sector' and calculate the sum of 'Amount($)' for each sector
#Sector_totals = df.groupby('Sector')['Amount($)'].sum().sort_values(ascending=False)
Amounts = df.groupby('Age')['Amount($)'].sum().reset_index()

Amount_totals = Amounts.sort_values(by='Age', ascending=False)

# Extract the data for plotting
Funding_mount = Amount_totals['Amount($)']
Age = Amount_totals['Age']
Amount_totals
```

```python
#Test Hypothesis with Kruskal-Wallis Test

# Group data by

import scipy.stats as stats  
grouped_data = df.groupby('HeadQuarter')['Amount($)'].apply(list)

# Perform Kruskal-Wallis Test
result = stats.kruskal(*grouped_data)

# Interpret the results
alpha = 0.05  # Set the significance level
p_value = result.pvalue

if p_value < alpha:
    print("Reject the null hypothesis (H_o). The location of a business affects the amount of funding received.")
else:
    print("Fail to reject the null hypothesis (H_o). The location of a business does not affect the amount of funding received.")
```
