### 4.3 BIVARIATE ANALYSIS

```python
# now lets have a look at the relationship between sector and amount raised
plt.figure(figsize=(10, 8))
plt.xticks(rotation=80) 
sns.boxplot(x=df['Sector'], y=df['Amount($)'], data=df)
plt.title('Bivariate Analysis of Raised Amount and Sector')
plt.show()
```

### we can see that the raised amount for most sectors is below $ 4000000 except for Finance that has an amount raised equivalent to about $5000000.

```python
# now lets have a look at the relationship between stage and amount raised
plt.figure(figsize=(8, 6))
plt.xticks(rotation=80) 
sns.boxplot(x=df['Stage'], y=df['Amount($)'], data=df)
plt.title('Bivariate Analysis of Raised Amount and Stage')
plt.show()
```

### we can see that amount raised is steadly increasing as the startup grows with highest amount is raised at the series C stage. we can deduce that at this stage a startup has likely achieved significant growth and success, which makes it more appealing to investors.

```python
# now lets have a look at the relationship between location and amount raised
plt.figure(figsize=(8, 6))
plt.xticks(rotation=80) 
sns.boxplot(x=df['HeadQuarter'], y=df['Amount($)'], data=df)
plt.title('Bivariate Analysis of Raised Amount and location')
plt.show()
```

### the maximum amount raised in most cities is below $4000000, while the lowest amount is about $1000000

```python
# now lets look at the relationship between stage of funding and startup sector
# Create a cross-tabulation (contingency table)
cross_table = pd.crosstab(df['Sector'], df['Stage'])

# Display the cross-tabulation
print(cross_table)
```

```python
# Create a heatmap of the cross-tabulation
plt.figure(figsize=(12, 8))
sns.heatmap(cross_table, cmap='YlGnBu', annot=True, fmt='d')
plt.title('Bivariate Analysis: Stage vs. Sector')
plt.xlabel('Sector')
plt.ylabel('Investor')
plt.show()
```

### we can see that in most sectors, startups are funded at the seed stage, which is an early stage of funding that typically comes after the initial startup capital, which often comes from the founders' own savings or contributions from friends and family. we can also see most sectors are funded at the series A stage where the startup has demonstrated some level of success and progress, proving its business model and gaining some traction in the market.
